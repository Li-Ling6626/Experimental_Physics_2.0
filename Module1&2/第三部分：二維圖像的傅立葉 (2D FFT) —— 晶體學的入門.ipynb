```python
# 引入本次實驗所需模組
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1.單狹縫與雙狹縫模擬
## 建立 2D 空間
N = 512
img = np.zeros((N, N))
```

```python
# 單狹縫
slit_width = 20

img[:, N//2 - slit_width//2 : N//2 + slit_width//2] = 1
```

```python
# 先畫單狹縫
plt.imshow(img, cmap="gray")
plt.title("Single Slit")
plt.axis("off")
plt.show()
```

```python
# 2D FFT
fft = np.fft.fftshift(np.fft.fft2(img))
intensity = np.abs(fft)**2
```

```python
# 畫出繞射圖樣
plt.imshow(np.log(intensity+1), cmap="inferno")
plt.title("Diffraction Pattern")
plt.axis("off")
plt.show()
```
