```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact
```

```python
N = 15
a = 1.0
E0 = 0
A = 1

m_vals = np.arange(-N//2, N//2)
k_vals = (2 * np.pi * m_vals) / (N * a)
```

```python
# 一維tight-binding
def E(k):
    return E0 - A * np.cos(k * a)
 
def plot_state(m_index):
    m = m_vals[m_index]
    k = k_vals[m_index]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
axs[0].scatter(k_vals, E(k_vals))
axs[0].scatter(k, E(k), s=100)
axs[0].set_xlabel("k")
axs[0].set_ylabel("E(k)")
axs[0].set_title("Discrete E-k")


sites = np.arange(N)
phase = np.exp(1j * k * sites * a)


axs[1].scatter(sites, np.real(phase), label="Re")
axs[1].scatter(sites, np.imag(phase), label="Im")
axs[1].set_xlabel("Site")
axs[1].set_ylabel("Wavefunction")
axs[1].set_title(f"m={m}, k={k:.2f}")
axs[1].legend()


plt.tight_layout()
plt.show()


interact(plot_state, m_index=(0, len(m_vals)-1))
```
